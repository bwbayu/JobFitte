# Merge training dataset

```python
def sample_or_all(group):
    return group.head(40)

# Apply the function to each group
df_jd = df_job.groupby('category').apply(sample_or_all).reset_index(drop=True)

print(df_jd['category'].info())

# Apply the function to each group
df_res = df_resume.groupby('category').apply(sample_or_all).reset_index(drop=True)

print(df_res['category'].info())
```

```python
# Rename columns for clarity
df_jd1 = df_jd.rename(columns={'clean_data': 'clean_jd'})
df_cv1 = df_res.rename(columns={'clean_data': 'clean_cv'})

# Create all possible pairs of job descriptions and CVs
df_pairs1 = pd.merge(df_cv1.assign(key=1), df_jd1.assign(key=1), on='key').drop('key', axis=1)

# Filter pairs where categories match
df_pairs1 = df_pairs1[df_pairs1['category_x'] == df_pairs1['category_y']]

# Assign labels
df_pairs1['label'] = 1

# Rename columns for clarity
df_pairs1 = df_pairs1.rename(columns={'category_x': 'category_cv', 'category_y': 'category_jd'})

# Select the desired columns
df_label1 = df_pairs1[['clean_cv', 'clean_jd', 'label']]

df_label1.head()
```

```python
df_label1.info()
```

```python
def sample_or_all(group):
    return group.head(7)

# Apply the function to each group
df_jd = df_job.groupby('category').apply(sample_or_all).reset_index(drop=True)

print(df_jd['category'].info())

# Apply the function to each group
df_res = df_resume.groupby('category').apply(sample_or_all).reset_index(drop=True)

print(df_res['category'].info())
```

```python
# Rename columns for clarity
df_jd2 = df_jd.rename(columns={'clean_data': 'clean_jd'})
df_cv2 = df_res.rename(columns={'clean_data': 'clean_cv'})

# Create all possible pairs of job descriptions and CVs
df_pairs2 = pd.merge(df_cv2.assign(key=1), df_jd2.assign(key=1), on='key').drop('key', axis=1)

# Filter pairs where categories match
df_pairs2 = df_pairs2[df_pairs2['category_x'] != df_pairs2['category_y']]

# Assign labels
df_pairs2['label'] = 0

# Rename columns for clarity
df_pairs2 = df_pairs2.rename(columns={'category_x': 'category_cv', 'category_y': 'category_jd'})

# Select the desired columns
df_label0 = df_pairs2[['clean_cv', 'clean_jd', 'label']]

df_label0.head()
```
